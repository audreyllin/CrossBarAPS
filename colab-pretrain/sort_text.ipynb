```python
def split_and_sort_text(text):
    # Split text into segments based on double line breaks
    segments = [segment.strip() for segment in text.split('\n\n') if segment.strip()]

    # Sort segments first by word count, then alphabetically
    sorted_segments = sorted(
        segments,
        key=lambda x: (len(x.split()), x.lower())  # (word count, text)
    )

    # Join them back with double line breaks
    return '\n\n'.join(sorted_segments)
```
