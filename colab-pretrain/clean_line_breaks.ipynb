```python
def clean_line_breaks(text):
    # Replace double line breaks with a temporary marker
    temp_marker = "||DOUBLE_LINE_BREAK||"
    text = text.replace('\n\n', temp_marker)

    # Remove all remaining single line breaks
    text = text.replace('\n', ' ')

    # Restore the double line breaks
    text = text.replace(temp_marker, '\n\n')

    return text

# Example usage
sample_text = """This is a paragraph with a single line break.
But it should continue on the same line.

This should be a new paragraph though.

And this is another paragraph."""

cleaned_text = clean_line_breaks(sample_text)
print(cleaned_text)
```
